```python
import json
def read_json(path):
    with open(path, "r") as f:
        coords = json.load(f)
    return coords
```

```python
import os
json_path = "../data/output/json"
json_name = "wonjun.json"
path = os.path.join(json_path, json_name)
frame_wise_coords = read_json(path)
```

```python
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
matplotlib.use("TkAgg")
%matplotlib tk
```

```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML    
import numpy as np

fig = plt.figure()
sub = fig.add_subplot(projection='3d')
sub.set_xlabel('X Label')
sub.set_ylabel('Y Label')
sub.set_zlabel('Z Label')

def update(frame):
    sub.clear()
    scene = frame_wise_coords[str(frame)]
    pose = scene['pose']
    landmark = []
    
    for i, lm in pose.items():
        landmark.append([lm['x'], lm['y'], lm['z']])
    landmark = np.array(landmark)
    
    xs, ys, zs = landmark[:, 0], landmark[:,1], landmark[:,2]
    sub.scatter(xs, ys, zs)

    
ani = FuncAnimation(fig, update, frames=range(1, len(frame_wise_coords)))

    
    
```
